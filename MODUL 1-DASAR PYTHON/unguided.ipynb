````python
def ambil_urut_ganjil(list1, list2):
    # Mengambil elemen dari indeks ganjil dari kedua list
    hasil = [list1[i] for i in range(1, len(list1), 2)] + [list2[i] for i in range(1, len(list2), 2)]
    
    # Mengurutkan hasil secara menurun
    hasil.sort(reverse=True)
    
    return hasil

# Contoh penggunaan fungsi
list1 = [13, 17, 21, 27, 31,39]
list2 = [7, 15, 27, 29, 43, 51]

hasil = ambil_urut_ganjil(list1, list2)
print(hasil)
````

````python
import csv

# Fungsi untuk membaca file CSV dan menyimpan data ke dalam dictionary
def baca_data_csv(nama_file):
    data_mahasiswa = {}
    try:
        with open(nama_file, mode='r') as file:
            reader = csv.reader(file)
            next(reader)  # Lewati header
            for row in reader:
                nama = row[0]
                nilai = int(row[1])  # Mengubah nilai dari string ke integer
                data_mahasiswa[nama] = nilai
    except FileNotFoundError:
        print(f"File {nama_file} tidak ditemukan.")
    except Exception as e:
        print(f"Terjadi kesalahan saat membaca file: {e}")
    return data_mahasiswa

# Fungsi untuk menampilkan mahasiswa dengan nilai tertinggi dan terendah
def tampilkan_tertinggi_terendah(data_mahasiswa):
    if not data_mahasiswa:  # Cek jika dictionary kosong
        print("Data mahasiswa kosong.")
        return
    
    mahasiswa_tertinggi = max(data_mahasiswa, key=data_mahasiswa.get)
    mahasiswa_terendah = min(data_mahasiswa, key=data_mahasiswa.get)
    
    print(f"Mahasiswa dengan nilai tertinggi: {mahasiswa_tertinggi} dengan nilai {data_mahasiswa[mahasiswa_tertinggi]}")
    print(f"Mahasiswa dengan nilai terendah: {mahasiswa_terendah} dengan nilai {data_mahasiswa[mahasiswa_terendah]}")

# Fungsi utama untuk menjalankan program
def main():
    nama_file = 'siswa_nilai (1).csv'  # Ganti dengan nama file CSV 
    data_mahasiswa = baca_data_csv(nama_file)  # Memanggil fungsi 
    
    # Tampilkan nilai tiap mahasiswa
    print("Nilai tiap mahasiswa:")
    for nama, nilai in data_mahasiswa.items():
        print(f"{nama}: {nilai}")
    
    print()
    
    # Tampilkan mahasiswa dengan nilai tertinggi dan terendah
    tampilkan_tertinggi_terendah(data_mahasiswa)

if __name__ == "__main__":
    main()
````

````python
import random

def tebak_angka():
    # Komputer memberikan pilihan angka acak antara 1 hingga 100
    angka_rahasia = random.randint(1, 100)
    
    # Total percobaan
    percobaan = 4
    
    print("Yuk tebak angka!")
    print("Aku kasih pilihan angka 1 hingga 100 yaa.")
    print(f"Kamu punya {percobaan} kesempatan untuk menebak angka tersebut.")
    
    # Mulai permainan
    for i in range(percobaan):
        try:
            tebakan = int(input(f"Tebakan {i+1}: Masukkan angka: "))
            
            if tebakan < 1 or tebakan > 100:
                print("Angka berada antara 1 hingga 100!")
                continue
            
            # Cek apakah tebakan benar
            if tebakan == angka_rahasia:
                print("Yeay! Tebakanmu benar!")
                return
            elif tebakan < angka_rahasia:
                print("Tebakanmu kecil banget.")
            else:
                print("Tebakanmu kebesaran.")
        
        except ValueError:
            print("Masukkan angka yang valid!")
    
    # Jika pengguna tidak berhasil menebak
    print(f"Sorry, masa percobaanmu udah abis. Angka yang benar adalah {angka_rahasia}.")

# Memulai permainan
tebak_angka()
````

````python
def faktorial(n):
    # Basis: jika n <= 1, kembalikan 1
    if n <= 1:
        return 1
    # Rekursi: n * faktorial(n-1)
    else:
        return n * faktorial(n - 1)

def urutan_faktorial(n):
    # List penyimpanan urutan faktorial
    result = []
    # Menghitung faktorial dari 1 hingga n
    for i in range(1, n + 1):
        result.append(faktorial(i))
    return result

# Penggunaan
n = 4
print("Input:", n)
print("Output:", urutan_faktorial(n))
````

````python
def min_coin_change(coins, amount):
    # Array untuk menyimpan jumlah minimum koin
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Jumlah uang 0 membutuhkan 0 koin

    # Loop melalui setiap jumlah dari 1 hingga amount
    for i in range(1, amount + 1):
        # Cek setiap koin yang tersedia
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    # Jika dp[amount] masih inf, maka tidak ada solusi
    if dp[amount] == float('inf'):
        return -1  # Tidak ada solusi
    else:
        return dp[amount]

# Fungsi untuk mencetak kombinasi koin yang digunakan
def find_coins_used(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    coin_used = [-1] * (amount + 1)  # Menyimpan koin yang digunakan untuk setiap jumlah

    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0 and dp[i - coin] + 1 < dp[i]:
                dp[i] = dp[i - coin] + 1
                coin_used[i] = coin

    if dp[amount] == float('inf'):
        return []

    result = []
    while amount > 0:
        result.append(coin_used[amount])
        amount -= coin_used[amount]

    return result

# Input dari pengguna
coins = list(map(int, input("Masukkan nilai koin yang tersedia (pisahkan dengan spasi): ").split()))
amount = int(input("Masukkan jumlah uang: "))

# Menentukan jumlah minimum koin
min_coins = min_coin_change(coins, amount)
if min_coins == -1:
    print(f"Tidak mungkin mencapai jumlah {amount} dengan koin yang tersedia.")
else:
    print(f"Jumlah minimum koin untuk mencapai {amount}: {min_coins}")
    # Menampilkan kombinasi koin yang digunakan
    coins_used = find_coins_used(coins, amount)
    print(f"Kombinasi koin yang digunakan: {coins_used}")
5
````

````python
def balik_kata(input_str):
    # Pisahkan string menjadi daftar kata-kata
    kata_list = input_str.split()
    
    # Balik setiap kata
    kata_terbalik = [kata[::-1] for kata in kata_list]
    
    return kata_terbalik

# Input  pengguna
input_str = input("Masukkan kalimat: ")

# Mengonversi dan menampilkan hasil
output = balik_kata(input_str)
print("Output:", output)

````

````python
from datetime import datetime

class Buku:
    # Inisialisasi atribut
    def __init__(self, judul, penulis, tahun_terbit):
        self.judul = judul
        self.penulis = penulis
        self.tahun_terbit = tahun_terbit

    # Method untuk menampilkan informasi buku
    def tampilkan_informasi(self):
        print(f"Judul: {self.judul}")
        print(f"Penulis: {self.penulis}")
        print(f"Tahun Terbit: {self.tahun_terbit}")
    
    # Method untuk menghitung usia buku
    def hitung_usia(self):
        tahun_sekarang = datetime.now().year
        usia = tahun_sekarang - self.tahun_terbit
        return usia

# Membuat 3 objek dari class Buku
buku1 = Buku("Atomic Habits", "James Clear", 2018)
buku2 = Buku("Filosofi Teras", "Henry Manampiring", 2018)
buku3 = Buku("The Alpha Girl's Guide", "Henry Manampiring", 2016)

# Menampilkan informasi dan usia masing-masing buku
for buku in [buku1, buku2, buku3]:
    buku.tampilkan_informasi()
    print(f"Usia Buku: {buku.hitung_usia()} tahun\n")
````

````python
def binary_search_even(arr, target):
    # Cek apakah target adalah angka ganjil
    if target % 2 != 0:
        return f"Nilai {target} tidak ada karena merupakan angka ganjil."
    
    # Inisialisasi batas bawah dan atas
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        # Jika target ditemukan di posisi tengah
        if arr[mid] == target:
            return f"Nilai {target} ditemukan di indeks {mid}."
        
        # Jika target lebih besar dari nilai tengah, periksa bagian kanan
        elif arr[mid] < target:
            low = mid + 1
        
        # Jika target lebih kecil dari nilai tengah, periksa bagian kiri
        else:
            high = mid - 1

    return f"Nilai {target} tidak ditemukan di dalam list."

# Contoh list yang berisi hanya angka genap
arr = [4, 8, 10, 18, 22, 28, 30, 38, 44, 56, 60]

# Input dari pengguna
target = int(input("Masukkan nilai yang ingin dicari: "))

# Melakukan pencarian biner
result = binary_search_even(arr, target)
print(result)
````
